```python
# Treating missing values

for i in df.columns:
    if df[i].dtypes=='object':
        m=df[i].mode()[0]
        df[i]=df[i].fillna(m)
    else:
        meanvalue=df[i].mean()
        df[i]=df[i].fillna(meanvalue)
```

```python
df.isna().sum()
```

```python
# Converting data types ( Timestamp    datetime64[ns]) change to datetime

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df
```

```python
df1=df.groupby('age').agg({'Customer_Reviews_Importance': 'sum', 'Rating_Accuracy ': 'sum','Shopping_Satisfaction':'sum'})
df1
```

```python
# Plot time series data using line plot

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,10))
sns.lineplot(data=df1, x='age', y='Customer_Reviews_Importance', label='Customer_RI')
sns.lineplot(data=df1, x='age', y='Rating_Accuracy ', label='Rating')
sns.lineplot(data=df1, x='age', y='Shopping_Satisfaction', label='Shopping')
plt.title('Amazon Customer Time Series Analysis')
plt.xlabel('Age')
plt.ylabel('Types')
plt.show()
```
